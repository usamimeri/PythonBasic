# concurrent.futures中的ThreadPoolExecutor 
其本质是对threading库的抽象，要添加更多控制还是建议用threading

ThreadPoolExecutor 是 Python 的线程池实现。它可以用于执行大量的异步任务,而不需要过于频繁的创建线程和销毁线程。
主要方法有:
- ThreadPoolExecutor(max_workers) : 初始化线程池,max_workers 是最大线程数。
- submit(fn, *args, **kwargs): 向线程池提交一个任务,返回一个 Future 对象。
- map(func, *iterables): 使用线程池执行func函数于iterables中的每个元素,返回结果列表。
- shutdown(): 关闭线程池,防止再提交新的任务。
- result(): 获取Future对象的结果,如果任务还没有完成会被阻塞。

# Threading模块


```python
from threading import Thread
import requests
def fetch(n):
    print(n)
    requests.get('http://httpbin.org/get',timeout=5)
def main():
    tasks = [Thread(target=fetch, args=(num,)) for num in range(20)]
    for task in tasks:
        task.start()
    for task in tasks:
        task.join()
 
if __name__ == '__main__':
    main()
```
